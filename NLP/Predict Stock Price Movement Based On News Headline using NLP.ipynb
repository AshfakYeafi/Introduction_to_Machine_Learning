```python
import pandas as pd
```

```python
df=pd.read_csv('stock_sentiment.csv',encoding='ISO-8859-1')
```

```python
df.head()
```

```python
train=df[df['Date']<'20150101']
test=df[df['Date']>'20141231']
```

```python
data=train.iloc[:,2:]
data.replace('[^a-zA-Z]',' ',regex=True,inplace=True)
```

```python
data
```

```python
list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
data.columns=new_index
data.head()
```

```python
for index in new_index:
    data[index]=data[index].str.lower()
data.head()
```

```python
headline=[]
```

```python
for row in range(0,len(data.index)):
    headline.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headline[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
cv=CountVectorizer(ngram_range=(2,2))
traindataset=cv.fit_transform(headline)
```

```python
rm=RandomForestClassifier(n_estimators=200,criterion='entropy')
```

```python
rm.fit(traindataset,train['Label'])
```

```python
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,0:25]))
test_dataset=cv.transform(test_transform)
predction=rm.predict(test_dataset)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predction)
print(matrix)
```

```python
score=accuracy_score(test['Label'],predction)
score
```

```python
report=classification_report(test["Label"],predction)
print(report)
```
